```python
# @title üìß Testador de SMTP Gmail { display-mode: "form" }
# @markdown Preencha os campos abaixo e execute a c√©lula (Play).

# @markdown **Importante:** Voc√™ deve cadastrar sua senha de app nos "Secrets" (√≠cone de chave üîë) com o nome `GMAIL_APP_PASS`.
# --- CAMPOS COM PLACEHOLDER ---
# @markdown Digite os e-mails nos campos abaixo:
seu_email = "luizhanauer@gmail.com" # @param {type:"string"}
destinatario = "luizhanauer@gmail.com" # @param {type:"string"}

import os
import time
import smtplib
from email.message import EmailMessage
from datetime import datetime
import pytz

os.environ['TZ'] = 'America/Sao_Paulo'
time.tzset()
print(f"Hor√°rio do Sistema: {datetime.now().strftime('%H:%M:%S')}")

try:
    # No Colab, isso pega o fuso do servidor, mas podemos for√ßar o nosso:
    fuso = pytz.timezone('America/Sao_Paulo')
    agora = datetime.now(fuso).strftime("%d/%m/%Y %H:%M:%S")
except:
    agora = datetime.now().strftime("%d/%m/%Y %H:%M:%S")

try:
    from google.colab import userdata
    HAS_COLAB = True
except ImportError:
    HAS_COLAB = False

# Valida√ß√£o simples para n√£o rodar vazio
if not seu_email or not destinatario:
    print("‚ö†Ô∏è AGUARDANDO DADOS: Por favor, preencha o seu e-mail e o destinat√°rio nos campos acima.")
else:
    agora = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    assunto = f"Teste de Conex√£o SMTP - {agora}"

    try:
        senha_app = userdata.get('GMAIL_APP_PASS') if HAS_COLAB else ""

        if not senha_app:
            raise ValueError("Chave 'GMAIL_APP_PASS' n√£o encontrada nos Secrets (√≠cone da chave üîë).")

        msg = EmailMessage()
        msg['Subject'] = assunto
        msg['From'] = seu_email
        msg['To'] = destinatario
        msg.set_content(f"E-mail de teste enviado com sucesso!\nDisparo: {agora}")

        print(f"[{agora}] üîÑ Conectando ao servidor Gmail...")

        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
            smtp.login(seu_email, senha_app)
            smtp.send_message(msg)

        print(f"[{agora}] ‚úÖ SUCESSO: E-mail enviado para {destinatario}!")

    except Exception as e:
        print(f"[{agora}] ‚ùå ERRO: {e}")
```
